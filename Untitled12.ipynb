```python
from sklearn.datasets import make_blobs
import numpy as np
from scipy.spatial.distance import cdist

class KMeansScratch:
    def __init__(self, k, max_iter=100):
        self.k = k
        self.max_iter = max_iter
        self.centroids = None

    def _initialize_centroids(self, X):
        # Randomly select k data points as initial centroids
        indices = np.random.choice(len(X), self.k, replace=False)
        return X[indices]

    def _assign_clusters(self, X):
        distances = cdist(X, self.centroids)
        return np.argmin(distances, axis=1)

    def _update_centroids(self, X, labels):
        new_centroids = np.array([X[labels == i].mean(axis=0) if np.any(labels == i) else self.centroids[i]
                                  for i in range(self.k)])
        return new_centroids

    def fit(self, X):
        self.centroids = self._initialize_centroids(X)

        for _ in range(self.max_iter):
            old_centroids = self.centroids
            labels = self._assign_clusters(X)
            self.centroids = self._update_centroids(X, labels)

            if np.allclose(self.centroids, old_centroids):
                break
        return labels, self.centroids

# The `silhouette_score_manual` and `visualize_clusters` functions are already defined in other cells.

X, y = make_blobs(n_samples=500, centers=3, random_state=42)

for k in [2, 3, 4]:
    print(f"\nRunning K-Means for K = {k}")

    model = KMeansScratch(k=k)
    labels, centroids = model.fit(X)

    score = silhouette_score_manual(X, labels)
    print(f"Silhouette Score = {score}")

    visualize_clusters(X, labels, centroids, k)
```
